```python
### Men
#---------------------------------------------------------------------------------------------------
## Cancer registry data
#c_A_M = pp.excel_extract(filesname_men[0], keyword = ['År','all'], num_parameter = num_region, 
                         #case = 'Tilfeller', cancer_type = 'Anus')
#c_MO_M = pp.excel_extract(filesname_men[0], keyword = ['År','all'], num_parameter = num_region, 
                          #case = 'Tilfeller', cancer_type = 'Munn, andre')
## incidance rate cases
incidence_anus_all = pp.excel_extract(filesname_men[0], keyword = ['År','all'], num_parameter = num_region, 
                                      case = 'Insidensrate', cancer_type = 'Anus')
incidence_other_all = pp.excel_extract(filesname_men[0], keyword = ['År','all'], num_parameter = num_region, 
                                       case = 'Insidensrate', cancer_type = 'Munn, andre')

## Separating norway data and region data
#c_A_regions = []
#c_MO_regions = []

# incidence rate 
inci_A_regions = []
inci_MO_regions = []

for i in range(num_years):
    #c_A_regions.append(c_anus_all[i][1:])
    #c_MO_regions.append(c_other_all[i][1:])
    # Calculating incidence rate 
    inci_A_regions.append( incidence_anus_all[i][1:] )
    inci_MO_regions.append( incidence_other_all[i][1:] )

inci_A_M = inci_A_regions
inci_MO_M = inci_MO_regions
#---------------------------------------------------------------------------------------------------

### Women
#---------------------------------------------------------------------------------------------------
## Cancer registry data
#c_livmorhals_W = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region,
#                                  case = 'Tilfeller', cancer_type = 'Livmorhals')

#c_A_W = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region, 
#                         case = 'Tilfeller', cancer_type = 'Anus')
#c_MO_W = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region, 
#                          case = 'Tilfeller', cancer_type = 'Munn, andre')

#c_livmorlegeme_W = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region,
#                                    case = 'Tilfeller', cancer_type = 'Livmorlegeme')
#c_livmor_W = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region,
#                              case = 'Tilfeller', cancer_type = 'Livmor, usesifisert')

## incidance rate cases
incidence_livmorhals_all = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region,
                                     case = 'Insidensrate', cancer_type = 'Livmorhals')

incidence_A_all = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region, 
                            case = 'Insidensrate', cancer_type = 'Anus')
incidence_MO_all = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region, 
                             case = 'Insidensrate', cancer_type = 'Munn, andre')

incidence_livmorlegeme_all = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region,
                                       case = 'Insidensrate', cancer_type = 'Livmorlegeme')
incidence_livmor_all = pp.excel_extract(filesname_women[0], keyword = ['År','all'], num_parameter = num_region,
                                 case = 'Insidensrate', cancer_type = 'Livmor, usesifisert')

## Separating norway data and region data
#c_livmorhals_regions = []
#c_A_regions = []
#c_MO_regions = []
#c_livmorlegeme_regions = []
#c_livmor_regions = []

#c_A_norway = []
#c_MO_norway = []

# incidence rate 
inci_livmorhals_regions = []
inci_A_regions = []
inci_MO_regions = []
inci_livmorlegeme_regions = []
inci_livmor_regions = []


for i in range(num_years):
    #c_livmorhals_regions.append( c_livmorhals_all[i][1:] )
    #c_A_regions.append( c_anus_all[i][1:] )
    #c_MO_regions.append( c_other_all[i][1:] )
    #c_livmorlegeme_regions.append( c_livmorlegeme_all[i][1:] )
    #c_livmor_regions.append( c_livmor_all[i][1:] )
    
    ## Calculating incidence rate
    inci_livmorhals_regions.append( incidence_livmorhals_all[i][1:] )
    inci_A_regions.append( incidence_anus_all[i][1:] )
    inci_MO_regions.append( incidence_other_all[i][1:] )
    inci_livmorlegeme_regions.append( incidence_livmorlegeme_all[i][1:] )
    inci_livmor_regions.append( incidence_livmor_all[i][1:] )

inci_livmorhals_W = inci_livmorhals_regions
inci_A_W = inci_A_regions
inci_MO_W = inci_MO_regions
inci_livmorlegeme_W = inci_livmorlegeme_regions
inci_livmor_W = inci_livmor_regions
#---------------------------------------------------------------------------------------------------
```

### HPV vaccine data

```python
## Men data preparation
#--------------------------------------------------------------------------------
dataframe = files_men[1][8:353]
pd_registry = dataframe

pd_registry['Year'] = dataframe['Unnamed: 1']
pd_registry['Gender'] = dataframe['Unnamed: 3']
pd_registry['Region'] = dataframe['Unnamed: 4']
pd_registry['Cases'] = dataframe['Unnamed: 6']

for i in range(1,9):
    # Dropping old columns
    pd_registry.drop(columns = ['Unnamed: '+str(i)+''], inplace = True)

# just some labels
pd_year_label = []
for i in range(4,10):
    pd_year_label.append( 'Year: 200'+str(i)+'' )
for i in range(10,19):
    pd_year_label.append( 'Year: 20'+str(i)+'' )
    
pd_divition = np.array(pd_registry['Region'][0:5])
pd_regions = np.array(pd_registry['Region'][5:23])

### extracting data
pd_data = []
for i in range(len(pd_year_label)):
    start = i*(len(pd_regions)+len(pd_divition))
    end = (1 + i)*(len(pd_regions)+len(pd_divition))
    
    pd_data.append( np.array(pd_registry['Cases'][start:end]) )

## changing 'under 5' to 2
pd_data2 = []
for i in range(len(pd_data)):
    for j in range(len(pd_data[i])):
        if pd_data[i][j] == 'under 5':
            pd_data[i][j] = 2
            
    pd_data2.append( pd_data[i] )

## separating divitions and regions
pd_casesD = []
pd_casesR = []
for i in range(len(pd_year_label)):
    pd_casesD.append( pd_data[i][0:5] )
    pd_casesR.append( pd_data[i][5:] )
    
pd_casesRM = pd_casesR
##-----------------------------------------------------------------------------

## Women data preparation
#--------------------------------------------------------------------------------
dataframe = files_women[1][8:353]
pd_registry = dataframe

pd_registry['Year'] = dataframe['Unnamed: 1']
pd_registry['Gender'] = dataframe['Unnamed: 3']
pd_registry['Region'] = dataframe['Unnamed: 4']
pd_registry['Cases'] = dataframe['Unnamed: 6']

for i in range(1,9):
    # Dropping old columns
    pd_registry.drop(columns = ['Unnamed: '+str(i)+''], inplace = True)

### extracting data
pd_data = []
for i in range(len(pd_year_label)):
    start = i*(len(pd_regions)+len(pd_divition))
    end = (1 + i)*(len(pd_regions)+len(pd_divition))
    
    pd_data.append( np.array(pd_registry['Cases'][start:end]) )

## changing 'under 5' to 2
pd_data2 = []
for i in range(len(pd_data)):
    for j in range(len(pd_data[i])):
        if pd_data[i][j] == 'under 5':
            pd_data[i][j] = 2
            
    pd_data2.append( pd_data[i] )

## separating divitions and regions
pd_casesD = []
pd_casesR = []
for i in range(len(pd_year_label)):
    pd_casesD.append( pd_data[i][0:5] )
    pd_casesR.append( pd_data[i][5:] )
    
pd_casesRW = pd_casesR
#--------------------------------------------------------------------------------
```

```python
### aligning Women's cancer and vaccine data
#---------------------------------------------------------------------------------
ordered_c_livmorhals_W = []
ordered_c_A_W = []
ordered_c_MO_W = []
ordered_c_livmorlegeme_W = []
ordered_c_livmor_W = []
ordered_v_W = []
## year loop
for i in range(len(labels)):
    temp_c_livmorhals = []
    temp_c_A = []
    temp_c_MO = []
    temp_c_livmorlegeme = []
    temp_c_livmor = []
    temp_v = []
    
    ## cancer loop
    for j in range(1,len(regions_xlabel)):
        ## vaccine loop
        for k in range(len(pd_regions)):
            # checking when they are equal
            if regions_xlabel[j] == pd_regions[k]:
                temp_c_livmorhals.append(inci_livmorhals_W[i][j-1])
                temp_c_A.append(inci_A_W[i][j-1])
                temp_c_MO.append(inci_MO_W[i][j-1])
                temp_c_livmorlegeme.append(inci_livmorlegeme_W[i][j-1])
                temp_c_livmor.append(inci_livmor_W[i][j-1])
                temp_v.append(pd_casesRW[i][k])
    
    ordered_c_livmorhals_W.append( np.array(temp_c_livmorhals) )
    ordered_c_A_W.append( np.array(temp_c_A) )
    ordered_c_MO_W.append( np.array(temp_c_MO) )
    ordered_c_livmorlegeme_W.append( np.array(temp_c_livmorlegeme) )
    ordered_c_livmor_W.append( np.array(temp_c_livmor) )
    ordered_v_W.append( np.array(temp_v) )
#---------------------------------------------------------------------------------

### aligning Men's cancer and vaccine data
#---------------------------------------------------------------------------------
ordered_c_A_M = []
ordered_c_MO_M = []
ordered_v_M = []
## year loop
for i in range(len(labels)):
    temp_c_A = []
    temp_c_MO = []
    temp_v = []
    
    ## cancer loop
    for j in range(1,len(regions_xlabel)):
        ## vaccine loop
        for k in range(len(pd_regions)):
            # checking when they are equal
            if regions_xlabel[j] == pd_regions[k]:
                temp_c_A.append(inci_A_M[i][j-1])
                temp_c_MO.append(inci_MO_M[i][j-1])
                temp_v.append(pd_casesRM[i][k])
    
    ordered_c_A_M.append( np.array(temp_c_A) )
    ordered_c_MO_M.append( np.array(temp_c_MO) )
    ordered_v_M.append( np.array(temp_v) )
#---------------------------------------------------------------------------------
```

```python
V_correlationM = []
V_correlationW = []

# running loop over all picked years 
for yp in range(len(ordered_v_M)):
    tempM = np.zeros(len(ordered_v_M[0]))
    tempW = np.zeros(len(ordered_v_M[0]))
    # running loop over all regions
    for a in range(len(ordered_v_M[0])):
        #running loop over all years under the picked year, namely yp
        for y in range(yp):
            tempM[a] = tempM[a] + ordered_v_M[y][a]
            tempW[a] = tempW[a] + ordered_v_M[y][a]
    
    V_correlationM.append( tempM )
    V_correlationW.append( tempW )
    
```

### Vaccine and cancer in the same plot (Total for each year)

```python
### Men
#-------------------------------------------------------------------------
temp1 = np.zeros(len(inci_MO_M))
temp2 = np.zeros(len(inci_MO_M))
temp3 = np.zeros(len(inci_MO_M))

for iyear in range(len(labels)):
    for iregion in range(len(inci_A_M[0])-1):
        temp1[iyear] = temp1[iyear] + pd_casesRM[iyear][iregion]
        temp2[iyear] = temp2[iyear] + inci_A_M[iyear][iregion]
        temp3[iyear] = temp3[iyear] + inci_MO_M[iyear][iregion]
    temp2[iyear] = temp2[iyear] + inci_A_M[iyear][iregion+1]
    temp3[iyear] = temp3[iyear] + inci_MO_M[iyear][iregion+1]
        
Mtotal_pd = temp1
Mtotal_cancer_A = temp2
Mtotal_cancer_MO = temp3


#-------------------------------------------------------------------------

### Women
#-------------------------------------------------------------------------
temp1 = np.zeros(len(inci_MO_W))
temp2 = np.zeros(len(inci_MO_W))
temp3 = np.zeros(len(inci_MO_W))
temp4 = np.zeros(len(inci_MO_W))
temp5 = np.zeros(len(inci_MO_W))
temp6 = np.zeros(len(inci_MO_W))
for iregion in range(len(pd_casesRW[0])-1):
    for iyear in range(len(pd_casesRW)-1):
        temp1[iyear] = temp1[iyear] + pd_casesRW[iyear][iregion]
        temp2[iyear] = temp2[iyear] + inci_livmorhals_W[iyear][iregion]
        temp3[iyear] = temp3[iyear] + inci_A_W[iyear][iregion]
        temp4[iyear] = temp4[iyear] + inci_MO_W[iyear][iregion]
        temp5[iyear] = temp5[iyear] + inci_livmorlegeme_W[iyear][iregion]
        temp6[iyear] = temp6[iyear] + inci_livmor_W[iyear][iregion]
        
Wtotal_pd = temp1
Wtotal_cancer_livmorhals = temp2
Wtotal_cancer_A = temp3
Wtotal_cancer_MO = temp4
Wtotal_cancer_livmorlegeme = temp5
Wtotal_cancer_livmor = temp6
```
